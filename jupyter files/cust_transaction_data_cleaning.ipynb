```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(r'.\data\Customer_Transaction_Data.csv')
data
```

```python
# ItemDesc values same as MerchCategoryDescription
data.drop(columns=['ItemDesc'], inplace=True)
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

```python
data[['SaleValue', 'OrderedQuantity']] = data[['SaleValue', 'OrderedQuantity']].abs()
```

```python
# Since only 9 records have nan we can drop them
data.dropna(subset=['SaleValue'], inplace=True)
```

```python
data.isna().sum()
```

```python
# Putting unknown as we can't assume its value
data['MerchCategoryDescription'].fillna('Unknown', inplace=True)
```

```python
# Since category value doesn't have repeated nan for class or group hence no point of using map() here
data[(data['MerchClassDescription'].isna()) & (data['MerchCategoryDescription'] != 'Unknown')]
```

```python
data[(data['MerchGroupDescription'].isna()) & (data['MerchCategoryDescription'] != 'Unknown')]
```

```python
# Fill na values with unknown for Class & Group columns
data['MerchClassDescription'].fillna('Unknown', inplace=True)
data['MerchGroupDescription'].fillna('Unknown', inplace=True)
```

```python
# To fill nan values in storeID using .map()
# Create a mapping from StoreCode to StoreID using non-null values
store_mapping = data.dropna(subset=['StoreID']).set_index('StoreCode')['StoreID']

# Convert mapping Series to a dictionary to deal with repeating values of StoreCode
store_mapping_dict = store_mapping.to_dict()

# Fill NaN values in StoreID column using the mapping
data['StoreID'] = data['StoreCode'].map(store_mapping_dict)
```

```python
# Using .map() again to store states based on city
city_mapping = data.dropna(subset=['StoreState']).set_index('StoreCity')['StoreState']

city_mapping_dict = city_mapping.to_dict()

data['StoreState'] = data['StoreCity'].map(city_mapping_dict)
```

```python
# Since we can't assume pincode values
data['StorePincode'].fillna('Unknown', inplace=True)
```

```python
# To clean pincodes with space like 400 097 = 400097
data['StorePincode'] = data['StorePincode'].str.replace(' ', '')
```

```python
# Since pincode value doesn't have nan for city or state hence no point of using pincode API
data[(data['StoreCity'].isna()) & (data['StorePincode'] != 'Unknown')]
```

```python
data[(data['StoreState'].isna()) & (data['StorePincode'] != 'Unknown')]
```

```python
# Since we can't use pincode values to fill City and State hence filling NaNs with unknown
data['StoreCity'].fillna('Unknown', inplace=True)
data['StoreState'].fillna('Unknown', inplace=True)
```

```python
(data['ReturnFlag'].isna().sum() / data['CustID'].count()) * 100
```

```python
# Since over 95 % values are NaN we'll set them as Unknown
data['ReturnFlag'].fillna('Unknown', inplace=True)
```
